```python
#['id', 'diamond_clarity', 'diamond_color', 'diamond_shape', 'carat_start', 'carat_end', 'factor']
df = pd.read_csv('factor.csv')

# fig = px.scatter_3d(df, x='diamond_color', y='diamond_clarity', z='factor',
#               opacity=0.7, color='factor',  symbol='carat_end')

# fig = px.scatter_3d(df, x='diamond_color', y='diamond_clarity', z='factor',
#               opacity=0.7, color='carat_end')

# fig = px.scatter_3d(df, x='diamond_color', y='diamond_clarity', z='factor',
#               opacity=0.7, color='carat_end')

fig = go.Figure(data=[go.Scatter3d(x=df.diamond_color,y=df.diamond_clarity, z=df.factor, opacity=0.8,
                                  marker=dict(size=3, color="#3266c1", symbol="circle", opacity=0.8))])
fig.show()
```
